### If, then, else

Write a program to ask students to input their numerical grade. The program will translate the numerical grade to A, B, C or Pass.

```python
# Inputing the grade
grade = input('Please input your grade: ')
```

```python
''' 
Write a if statement to 
- If the grade is >= 90, print 'A'.
- If the grade is >=80 and <90, print 'B'.
- If the grade is >=70 and <80, print 'C'.
- Otherwise, print 'Pass'.
'''
grade=int(grade)

if grade >= 90:
    print('A')
elif (grade >=80) and (grade <90):
    print('B')
elif (grade>=70) and (grade<80):
    print('C')
else:
    print('Pass')
```

### Function

A wrap of program that can be re-used without re-writing the program

```python
ls=[]
for i in range(0,1000):
    if (i % 13) == 0:
        ls.append(i)

print(sum(ls))
```

```python
 # Wrap the above program into a function num13(start_num,end_num), which calculate the sum of numbers that can be 

def num13(s_num,e_num):
    ls=[]
    for i in range(s_num,e_num):
        if (i % 13) == 0:
            ls.append(i)
    return sum(ls)

num13(1,11)
```

# 2. Numpy Array

- A numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers.
- An array can be 1-dimensional, 2-dimensional or n-dimensional. For instance, a raster is essentially a geo-coordinated 2D array.
- Arrays support element-wise operations that is not supported by other python data collections (e.g. list)

# 4. Loading Spatial Data

## 4.1 Vector data
Vector data represents discrete objects.

### Elements of vector data
- Geometries (points, lines and polygons)
- Attributes (names, population, temperature, income...)

### Structure of vector data
- Attributes are organized in tables
    - Each column is an attribute
    - Each row stores the attributes of a spatial object
- Each geometry is linked with a set of attributes (a row in the attribute table)
- You can consider geometries are a special column in the attribute table.

## 4.2 Raster data

### Element of raster data
- A 2D matrix
- Spatial reference (coordinate) system
- Top-left corner coordinates
- Cell-size (resolution)

<img src="images/raster.PNG" alt="Drawing" align="left" style="width: 800px;"/>

```python
# import needed packages
import rasterio as rio
import rasterio.plot as rio_pl

# enable interactive plotting
import matplotlib.pyplot as plt
plt.ion()
plt.rcParams['figure.figsize'] = [45, 45]
```

```python
# read the raster data (tif format) from the local drive
raster = rio.open("C:/Users/yi/Documents/UH_work/Teaching/GEOG389/other_materials/data/dem_oahu.tif")
```

```python
# Plot the raster
rio_pl.show(raster, cmap='OrRd')
```

```python
raster.profile
```

The `afines` property includes 6 numbers (a, b ,c, d, e, f)
